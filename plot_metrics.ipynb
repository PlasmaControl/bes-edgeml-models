```python
def get_lists(d: dict):
    scaled_pos = []
    precision = []
    recall = []
    f1 = []
    for key, val in d.items():
        for k, v in val.items():
            if k == "scaled_pos":
                scaled_pos.append(v)
            elif k == "precision":
                precision.append(v)
            elif k == "recall":
                recall.append(v)
            elif k == "f1":
                f1.append(v)
    return scaled_pos, precision, recall, f1
```
