```python
def calculate_error_rate(fp, fn, tp, tn):
    """
        error rate = (fp + fn) / (fp + fn + tp + tn)
    """
    return (fp+fn) / (fp+fn+tp+tn)

def calculate_accuracy(fp, fn, tp, tn):
    """
        accuracy = (tp + tn) / (fp + fn + tp + tn)
    """
    return (tp+tn) / (fp+fn+tp+tn)

def calculate_recall(fn, tp):
    """
        recall (sensitivity or TPR) = tp / (fn + tp)
    """
    return tp / (fn+tp)

def calculate_specificity(fp, tn):
    """
        specificity (TNR) = tn / (fp + tn)
    """
    return tn / (fp+tn)

def calculate_precision(tp, fp):
    """
        precision = tp / (tp + fp)
    """
    return tp / (tp + fp)
```
