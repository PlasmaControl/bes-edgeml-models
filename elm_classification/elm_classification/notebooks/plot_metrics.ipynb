```python
def get_lists(d: dict):
    scaled_pos = []
    precision = []
    recall = []
    f1 = []
    for key, val in d.items():
        for k, v in val.items():
            if k == "scaled_pos":
                scaled_pos.append(v)
            elif k == "precision":
                precision.append(v)
            elif k == "recall":
                recall.append(v)
            elif k == "f1":
                f1.append(v)
    return scaled_pos, precision, recall, f1
```

```python
base_dir = 'outputs/signal_window_64'
root_dir = os.path.dirname(os.getcwd())
base_path = os.path.join(root_dir, base_dir)
def get_roc_data(base_path:str, model_name: str, look_ahead:int=0):
    path = os.path.join(base_path, f'label_look_ahead_{look_ahead}', 'roc')
    csv_file_path = os.path.join(path, f'{model_name}_roc_details_micro_lookahead_{look_ahead}_wavelet.csv')
    df = pd.read_csv(csv_file_path)
    df = df.loc[1:, :]
    return df
```
