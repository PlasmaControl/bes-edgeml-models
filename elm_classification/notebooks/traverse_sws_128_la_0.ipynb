```python
path = 'elm_classification/outputs/signal_window_128/label_look_ahead_0/training_metrics/'
```

```python
%matplotlib inline
import os
import pickle

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', palette='muted', font_scale=1.15)
```

```python
pkl_fname = 'training_metrics_multi_features.pkl'
with open(os.path.join(path, pkl_fname), "rb") as f:
    metrics = pickle.load(f)
```

```python
metrics
```

```python
train_loss = metrics['train_loss']
valid_loss = metrics['valid_loss']
roc_scores = metrics['roc_scores']
f1_scores = metrics['f1_scores']
```

```python
fig = plt.figure(figsize=(8, 6), dpi=100)
num_epochs = list(range(1, len(train_loss)+1))
plt.plot(num_epochs, train_loss, label='train loss')
plt.xlabel('Epochs')
plt.ylabel('Train Loss')
plt.title('Signal window size: 128, label lookahead: 0')
plt.show()
```

```python
fig = plt.figure(figsize=(8, 6), dpi=100)
plt.plot(num_epochs, valid_loss, label='valid loss')
plt.xlabel('Epochs')
plt.ylabel('Valid Loss')
plt.title('Signal window size: 128, label lookahead: 0')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(8, 6), dpi=100)
plt.plot(num_epochs, roc_scores, label='ROC score')
plt.xlabel('Epochs')
plt.ylabel('ROC score')
plt.title('Signal window size: 128, label lookahead: 0')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(8, 6), dpi=100)
plt.plot(num_epochs, f1_scores, label='F1 score')
plt.xlabel('Epochs')
plt.ylabel('F1 score')
plt.title('Signal window size: 128, label lookahead: 0')
plt.legend()
plt.show()
```
